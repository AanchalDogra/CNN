```python
import seaborn as sns
from scipy import stats
```

```python
x_pop_50 = stats.norm(50,1).rvs(size = 1000) # normal disbn with mean = 50 and std deviation = 1 and rvs is random variable simulation of 1000 is created
```

```python
plt.hist(x_pop_50, bins = 50)
plt.show()
```

```python
x_pop_0 = stats.norm(0,1).rvs(size = 1000) # normal disbn with mean = 0 and std deviation = 1 and rvs is random variable simulation of 1000 is created
plt.hist(x_pop_0, bins = 50)
plt.axvline(np.mean(x_pop_0), color = 'k', linestyle = 'dashed',linewidth = 2)
plt.show()
```

```python
significance_level = 0.95
```

```python
t_critical = stats.norm(0,1).ppf(significance_level)
```

```python
stats.norm(50,6).ppf(significance_level)
```

```python
x_pop_0 = stats.norm(0,1).rvs(size = 1000) # normal disbn with mean = 0 and std deviation = 1 and rvs is random variable simulation of 1000 is created
plt.hist(x_pop_0, bins = 50)
plt.axvline(np.mean(x_pop_0), color = 'k', linestyle = 'dashed',linewidth = 2)
plt.axvline(t_critical, color = 'r', linestyle = 'dashed',linewidth = 2)

plt.show()
```
